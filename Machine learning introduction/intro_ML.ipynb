### Классификация задач
Задачи могут относиться к различным классам в зависимости от типа проблемы:

**Классификация** - Задача состоит в присвоении объекту одного из нескольких классов. Например, определить что за фрукт на изображении.

**Регрессия** - Задача заключается в предсказании числового значения. Например, определить стоимость того или иного товара в следующем году.

**Кластеризация** - Разделение объектов на группы(кластеры) на основе схожести их характеристик.

**Ассоциация** - Нахождение закономерностей или правил взаимосвязи между объектами.

**Понижение размерности** - Преобразование множества данных таким образом, чтобы уменьшить количество признаков, сохранив при этом основную информацию
- ***Системы для рекомендаций*** - классификация

- ***Распознование изображений*** - классификация

- ***Прогнозирование спроса*** - регрессия

- ***Фильтрация спама*** - классификация

- ***Медицинская диагностика*** - классификация

### Классификация задач из введения
- ***Спрогнозируйте цену дома*** - регрессия

- ***Предсказать, вернет ли клиент кредит*** - мультиклассовая классификация, с целью = зарплата и кредитная история

- ***Предсказать, когда пациенту необходимо принять лекарство*** - регрессия, если время приема лекарства можно представить как непрерывную величину (например, часы или минуты). В этом случае задача сводится к предсказанию числового значения времени. классификация,если время приема лекарства можно представить как дискретную величину (например, утро, день, вечер). В этом случае задача сводится к предсказанию категории времени.

- ***Выберите, какое лекарство из имеющихся в наличии следует принимать пациенту*** - мультиклассовая классификация, с целью = Текущие медицинские карты пациентов. Результаты рандомизированного контролируемого исследования с этим препаратом. А также медицинские карты других пациентов.

- ***Выберите сегмент клиентов для промо-коммуникации*** - классификация, понижение размерности, ассоциации, цель =
участвует ли клиент в акции или нет, покупает ли клиент товары, участвующие в акции

- ***Распознавание дефектной продукции на производственной линии (на основе сканирования фотографий)*** - классификация, цель = есть дефект или нет

- ***Решите, как разместить продукты на полке в магазине*** - кластеризация

- ***Поиск сайтов для ввода текстового запроса*** - классификация, потому что мы пытаемся отнести каждый сайт к одной из заранее определенных категорий. Цель = открывает ли пользователь ссылку, мультикласс — плохой вариант, так как количество страниц слишком велико.

- ***Разделите покупателей магазина на сегменты, чтобы понять различия в их поведении*** - кластеризация

- ***Обнаружение аномалий в трафике сайта*** - Кластеризация — если у нас нет размеченных данных и мы хотим автоматически обнаружить аномалии на основе внутренних закономерностей данных, или классификация, — если у нас есть размеченные данные с примерами нормального и аномального трафика. (если мы
обнаружили факты неисправностей)

### Разница между multiclass и multilabel классификацией:
**Multiclass классификация:** каждый образец может принадлежать только к одному из классов из множества возможных. Это задача классификации, состоящая из более чем двух Классы. Каждый образец может быть помечен только как один класс. Например, классификация с помощью признаков, извлеченных из набора изображений фрукт, где каждое изображение может быть апельсином, яблоком или грушей. Каждое изображение является одним образцом и помечено как один из 3 возможных классов. Многоклассовая классификация предполагает присвоение каждой выборки к одной и только одной этикетке - один образец не может, например, быть одновременно грушей и яблоко.

**Multilabel классификация:** один образец может принадлежать к нескольким классам. Например, фотография может содержать в себе человека, машину и лес.
Многоклассовая классификация присваивает каждому образцу набор целевых меток. Это можно рассматривать как прогнозирование свойств точки данных, которые не являются взаимоисключающими. Например, прогнозирование тем, релевантных текстовому документу или видео. Документ или видео может быть о «религии», «политике», «финансах» или «образование», несколько тематических классов или все тематические классы. Взято из https://scikit-learn.org/stable/modules/multiclass.html

### Является ли пример с ценами на жильё из теории классификацией задачи регрессии? Можно ли свести задачу регрессии к классификации?
Да, эта задача относится к задаче регрессии, так как требуется предсказание численного значения (цены жилья). Однако её можно свести к задаче классификации, если разбить диапазон возможных цен на интервалы и считать каждый интервал отдельным классом. Но такой подход обычно менее эффективен, поскольку теряется точность предсказаний. Теоретический пример — задача регрессии. Мы можем свести задачу регрессии к задаче классификации, стратифицировав нашу целевую переменную. Другими словами, можно разделить непрерывную целевую переменную на дискретные группы.
