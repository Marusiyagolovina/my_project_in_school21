## Naive regressor (–ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ —Å—Ä–µ–¥–Ω–µ–≥–æ —Ä–µ–π—Ç–∏–Ω–≥–∞)

```python
# –ù–∞–∏–≤–Ω—ã–π –∞–ª–≥–æ—Ä–∏—Ç–º 1: –ü—Ä–µ–¥—Å–∫–∞–∑—ã–≤–∞–µ–º —Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ
mean_rating = y_train.mean()
y_pred_mean = np.full_like(y_test, mean_rating)

# –ù–∞–∏–≤–Ω—ã–π –∞–ª–≥–æ—Ä–∏—Ç–º 2: –ü—Ä–µ–¥—Å–∫–∞–∑—ã–≤–∞–µ–º –º–µ–¥–∏–∞–Ω—É
median_rating = y_train.median()
y_pred_median = np.full_like(y_test, median_rating)

# –ù–∞–∏–≤–Ω—ã–π –∞–ª–≥–æ—Ä–∏—Ç–º 3: –ü—Ä–µ–¥—Å–∫–∞–∑—ã–≤–∞–µ–º —Å–∞–º–æ–µ —á–∞—Å—Ç–æ–µ –∑–Ω–∞—á–µ–Ω–∏–µ (–º–æ–¥—É)
mode_rating = y_train.mode()[0] if not y_train.mode().empty else y_train.median()
y_pred_mode = np.full_like(y_test, mode_rating)

# 5. –û—Ü–µ–Ω–∫–∞ –∫–∞—á–µ—Å—Ç–≤–∞ –Ω–∞–∏–≤–Ω—ã—Ö –∞–ª–≥–æ—Ä–∏—Ç–º–æ–≤
def evaluate_predictions(y_true, y_pred, model_name):
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_true, y_pred)
    
    
    print(f"\nüìà {model_name}:")
    print(f"MAE: {mae:.4f}")
    print(f"MSE: {mse:.4f}")
    print(f"RMSE: {rmse:.4f}")
    print(f"R¬≤: {r2:.4f}")
    
    return {'MAE': mae, 'MSE': mse, 'RMSE': rmse, 'R¬≤': r2}

# –û—Ü–µ–Ω–∏–≤–∞–µ–º –≤—Å–µ –Ω–∞–∏–≤–Ω—ã–µ –∞–ª–≥–æ—Ä–∏—Ç–º—ã
results = {}
results['Mean'] = evaluate_predictions(y_test, y_pred_mean, "–ù–∞–∏–≤–Ω—ã–π –∞–ª–≥–æ—Ä–∏—Ç–º (–°—Ä–µ–¥–Ω–µ–µ)")
results['Median'] = evaluate_predictions(y_test, y_pred_median, "–ù–∞–∏–≤–Ω—ã–π –∞–ª–≥–æ—Ä–∏—Ç–º (–ú–µ–¥–∏–∞–Ω–∞)")
results['Mode'] = evaluate_predictions(y_test, y_pred_mode, "–ù–∞–∏–≤–Ω—ã–π –∞–ª–≥–æ—Ä–∏—Ç–º (–ú–æ–¥–∞)")

# 7. –ê–Ω–∞–ª–∏–∑ –ª—É—á—à–µ–≥–æ –Ω–∞–∏–≤–Ω–æ–≥–æ –∞–ª–≥–æ—Ä–∏—Ç–º–∞
best_naive = min(results, key=lambda x: results[x]['RMSE'])
print(f"\nüéØ –õ—É—á—à–∏–π –Ω–∞–∏–≤–Ω—ã–π –∞–ª–≥–æ—Ä–∏—Ç–º: {best_naive}")
print(f"RMSE: {results[best_naive]['RMSE']:.4f}")
print(f"MAE: {results[best_naive]['MAE']:.4f}")

# 8. –ë–µ–Ω—á–º–∞—Ä–∫ –¥–ª—è —Å—Ä–∞–≤–Ω–µ–Ω–∏—è —Å –±—É–¥—É—â–∏–º–∏ –º–æ–¥–µ–ª—è–º–∏
print(f"\nüìä –ë–µ–Ω—á–º–∞—Ä–∫ - –Ω–∞–∏–≤–Ω—ã–π –∞–ª–≥–æ—Ä–∏—Ç–º ({best_naive}):")
print(f"–°—Ä–µ–¥–Ω—è—è –æ—à–∏–±–∫–∞: ¬±{results[best_naive]['MAE']:.3f} –±–∞–ª–ª–∞")
print(f"–¢–∏–ø–∏—á–Ω–∞—è –æ—à–∏–±–∫–∞: ¬±{results[best_naive]['RMSE']:.3f} –±–∞–ª–ª–∞")

# 9. –ê–Ω–∞–ª–∏–∑ –≤–∞–∂–Ω–æ—Å—Ç–∏ - –∫–∞–∫–∏–µ –∏–Ω–≥—Ä–µ–¥–∏–µ–Ω—Ç—ã —á–∞—â–µ –≤—Å—Ç—Ä–µ—á–∞—é—Ç—Å—è —É –≤—ã—Å–æ–∫–∏—Ö —Ä–µ–π—Ç–∏–Ω–≥–æ–≤
high_rated = df[df['rating'] >= df['rating'].quantile(0.75)]  # –¢–æ–ø 25% —Ä–µ—Ü–µ–ø—Ç–æ–≤
low_rated = df[df['rating'] <= df['rating'].quantile(0.25)]   # –ù–∏–∑ 25% —Ä–µ—Ü–µ–ø—Ç–æ–≤

# –°—Ä–∞–≤–Ω–µ–Ω–∏–µ —á–∞—Å—Ç–æ—Ç—ã –∏–Ω–≥—Ä–µ–¥–∏–µ–Ω—Ç–æ–≤
ingredient_impact = {}
for col in X.columns:
    if col != 'rating':
        high_freq = high_rated[col].mean()
        low_freq = low_rated[col].mean()
        impact = high_freq - low_freq
        ingredient_impact[col] = impact

# –¢–æ–ø 10 –∏–Ω–≥—Ä–µ–¥–∏–µ–Ω—Ç–æ–≤ —Å –Ω–∞–∏–±–æ–ª—å—à–∏–º –≤–ª–∏—è–Ω–∏–µ–º –Ω–∞ —Ä–µ–π—Ç–∏–Ω–≥
top_positive = sorted(ingredient_impact.items(), key=lambda x: x[1], reverse=True)[:10]
top_negative = sorted(ingredient_impact.items(), key=lambda x: x[1])[:10]

print(f"\n –¢–æ–ø 10 –∏–Ω–≥—Ä–µ–¥–∏–µ–Ω—Ç–æ–≤ –¥–ª—è –≤—ã—Å–æ–∫–æ–≥–æ —Ä–µ–π—Ç–∏–Ω–≥–∞:")
for ing, impact in top_positive:
    print(f"  {ing}: +{impact:.3f}")

print(f"\n –¢–æ–ø 10 –∏–Ω–≥—Ä–µ–¥–∏–µ–Ω—Ç–æ–≤ –¥–ª—è –Ω–∏–∑–∫–æ–≥–æ —Ä–µ–π—Ç–∏–Ω–≥–∞:")
for ing, impact in top_negative:
    print(f"  {ing}: {impact:.3f}")
```

#### –†–∞–∑–±–µ—Ä—ë–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã:
- Mean (—Å—Ä–µ–¥–Ω–µ–µ –∑–Ω–∞—á–µ–Ω–∏–µ —Ä–µ–π—Ç–∏–Ω–≥–∞) ‚Üí –ª—É—á—à–∏–π baseline:

    - MAE ‚âà 0.895

    - RMSE ‚âà 1.341

    - R¬≤ ‚âà 0.0 (–∫–∞–∫ –∏ –ø–æ–ª–æ–∂–µ–Ω–æ –Ω–∞–∏–≤–Ω–æ–π –º–æ–¥–µ–ª–∏).

- –ú–µ–¥–∏–∞–Ω–∞ –∏ –º–æ–¥–∞ —Ö—É–∂–µ ‚Äî –≤—ã—à–µ RMSE, –Ω–∏–∂–µ –æ–±—ä—è—Å–Ω—ë–Ω–Ω–∞—è –¥–∏—Å–ø–µ—Ä—Å–∏—è.

**–ó–Ω–∞—á–∏—Ç, –ª—é–±–∞—è –º–æ–¥–µ–ª—å –¥–æ–ª–∂–Ω–∞ –±–∏—Ç—å —ç—Ç–æ—Ç –ø–æ—Ä–æ–≥ (MAE < 0.895 –∏ RMSE < 1.341)**, –∏–Ω–∞—á–µ –æ–Ω–∞ —Ö—É–∂–µ "—É–≥–∞–¥—ã–≤–∞–Ω–∏—è —Å—Ä–µ–¥–Ω–µ–≥–æ".

üçΩÔ∏è **–ò–Ω–≥—Ä–µ–¥–∏–µ–Ω—Ç—ã –∏ —Ä–µ–π—Ç–∏–Ω–≥**

- –ü–æ–ª–æ–∂–∏—Ç–µ–ª—å–Ω–æ–µ –≤–ª–∏—è–Ω–∏–µ: –º—è—Å–æ, –º–æ–ª–æ—á–Ω—ã–µ –ø—Ä–æ–¥—É–∫—Ç—ã (—Å—ã—Ä, –º–æ–ª–æ–∫–æ, —Å–ª–∏–≤–∫–∏), –∫–æ–ª–±–∞—Å—ã, —è–≥–æ–¥—ã (–∫–ª—é–∫–≤–∞).

- –û—Ç—Ä–∏—Ü–∞—Ç–µ–ª—å–Ω–æ–µ –≤–ª–∏—è–Ω–∏–µ: –æ–≤–æ—â–∏ (–∫–∞–ø—É—Å—Ç–∞, –∫–∞–±–∞—á–æ–∫, —Ä–µ–¥–∏—Å, –±—Ä–æ–∫–∫–æ–ª–∏, –ø–µ—Ä–µ—Ü), —Ä–∏—Å, –±–∞–Ω–∞–Ω, —Ç–æ—Ñ—É, –π–æ–≥—É—Ä—Ç.

–≠—Ç–æ —É–∂–µ **–¥–∞—ë—Ç –≥–∏–ø–æ—Ç–µ–∑—É:**

- "—Å—ã—Ç–Ω—ã–µ" —Ä–µ—Ü–µ–ø—Ç—ã (–º—è—Å–æ+–º–æ–ª–æ—á–∫–∞) —á–∞—â–µ –ø–æ–ª—É—á–∞—é—Ç –≤—ã—Å–æ–∫–∏–µ –æ—Ü–µ–Ω–∫–∏;

- "–∑–¥–æ—Ä–æ–≤–∞—è –µ–¥–∞" –∏ –ø—Ä–æ—Å—Ç—ã–µ –≥–∞—Ä–Ω–∏—Ä—ã ‚Üí –Ω–∞–æ–±–æ—Ä–æ—Ç.



### –°—Ä–∞–≤–Ω–∏–º —Å baseline —Å–ª–µ–¥—É—é—â–∏–µ –º–æ–¥–µ–ª–∏:

## Linear Regression

```python
# 2. –ú–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏–µ
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# 3. –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏
model = LinearRegression()
model.fit(X_train_scaled, y_train)
```

```python
# 4. –ü—Ä–æ–≥–Ω–æ–∑ –∏ –æ—Ü–µ–Ω–∫–∞

y_pred = model.predict(X_test_scaled)

from sklearn.metrics import mean_squared_error, r2_score

mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f"MSE: {mse:.3f}, RMSE: {rmse:.3f}, R2: {r2:.3f}")
```

- –®—É–º–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ: –≤–æ–∑–º–æ–∂–Ω–æ, —Ä–µ–π—Ç–∏–Ω–≥–∏ —Å–æ–¥–µ—Ä–∂–∞—Ç –º–Ω–æ–≥–æ —Å–ª—É—á–∞–π–Ω–æ—Å—Ç–∏ (—Å—É–±—ä–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç—å –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–µ–π).

- –ú–∞–ª–æ –∏–Ω—Ñ–æ—Ä–º–∞—Ç–∏–≤–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤: –µ—Å–ª–∏ –≤—Å–µ —Ñ–∏—á–∏ ‚Äî –ø—Ä–æ—Å—Ç–æ 0/1, –º–æ–¥–µ–ª—å –≤–∏–¥–∏—Ç –æ—á–µ–Ω—å –æ–≥—Ä–∞–Ω–∏—á–µ–Ω–Ω—É—é –∫–∞—Ä—Ç–∏–Ω—É.

## XGBoost

```python
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, r2_score

# –ú–æ–¥–µ–ª—å XGBoost
xgb_model = XGBRegressor(
    n_estimators=500,      # –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –¥–µ—Ä–µ–≤—å–µ–≤
    learning_rate=0.05,    # —Å–∫–æ—Ä–æ—Å—Ç—å –æ–±—É—á–µ–Ω–∏—è
    max_depth=6,           # –≥–ª—É–±–∏–Ω–∞ –¥–µ—Ä–µ–≤—å–µ–≤
    subsample=0.8,         # –ø–æ–¥–≤—ã–±–æ—Ä–∫–∞
    colsample_bytree=0.8,  # –≤—ã–±–æ—Ä —Ñ–∏—á–µ–π
    random_state=21,
    n_jobs=-1
)

# –û–±—É—á–µ–Ω–∏–µ
xgb_model.fit(X_train, y_train)

# –ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è
y_pred = xgb_model.predict(X_test)

# –ú–µ—Ç—Ä–∏–∫–∏
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(f"MSE: {mse:.3f}, RMSE: {rmse:.3f}, R2: {r2:.3f}")
```

## –ù–∞—Å—Ç—Ä–æ–∏–º –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä—ã —á–µ—Ä–µ–∑ GridSearchCV

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import GridSearchCV, StratifiedKFold, train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
import time
import warnings

# –ü–æ–¥–∞–≤–ª—è–µ–º FutureWarning sklearn
warnings.filterwarnings("ignore", category=FutureWarning, module="sklearn")

# --- 1. –°–æ–∑–¥–∞–Ω–∏–µ –∫–ª–∞—Å—Å–æ–≤ ---
print("–°–æ–∑–¥–∞–Ω–∏–µ –∫–ª–∞—Å—Å–æ–≤...")
y_rounded = np.round(y).astype(int)
unique_classes = np.unique(y_rounded)
print(f"–£–Ω–∏–∫–∞–ª—å–Ω—ã–µ –∫–ª–∞—Å—Å—ã: {unique_classes} ({len(unique_classes)})")
print("\n–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–æ–≤:")
for r in sorted(unique_classes):
    count = sum(y_rounded == r)
    print(f"–†–µ–π—Ç–∏–Ω–≥ {r}: {count} ({count/len(y_rounded)*100:.1f}%)")

# --- 2. –†–∞–∑–¥–µ–ª–µ–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö ---
X_train_5class, X_test_5class, y_train_5class, y_test_5class = train_test_split(
    X, y_rounded, test_size=0.2, random_state=21, stratify=y_rounded
)
print(f"–†–∞–∑–º–µ—Ä train: {X_train_5class.shape}, test: {X_test_5class.shape}")

# --- 3. –ù–∞—Å—Ç—Ä–æ–π–∫–∞ GridSearch ---
cv_strategy = StratifiedKFold(n_splits=3, shuffle=True, random_state=21)

'''
# –†–∞—Å—à–∏—Ä–µ–Ω–Ω—ã–µ —Å–µ—Ç–∫–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –¥–ª—è 5 –∫–ª–∞—Å—Å–æ–≤
param_grids_5class = {
    'Logistic Regression': {
        'C': [0.01, 0.1, 1, 10],
        'solver': ['lbfgs'],
        'max_iter': [1000, 2000]
    },
    'Random Forest': {
        'n_estimators': [50, 100, 200],
        'max_depth': [None, 10, 20, 30],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4],
        'class_weight': ['balanced', None]
    },
    'XGBoost': {
        'n_estimators': [100, 200, 300],
        'learning_rate': [0.01, 0.05, 0.1, 0.2],
        'max_depth': [3, 5, 7, 9],
        'subsample': [0.6, 0.8, 1.0],
        'colsample_bytree': [0.6, 0.8, 1.0],
        'gamma': [0, 0.1, 0.2],
        'reg_alpha': [0, 0.1, 0.5],
        'reg_lambda': [1, 1.5, 2]
        # scale_pos_weight —É–±—Ä–∞–Ω –¥–ª—è –º–Ω–æ–≥–æ–∫–ª–∞—Å—Å–æ–≤–æ–π –∑–∞–¥–∞—á–∏
    }
}
'''

# –£–ø—Ä–æ—â–µ–Ω–Ω—ã–µ —Å–µ—Ç–∫–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –¥–ª—è –±—ã—Å—Ç—Ä–æ–≥–æ –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è
param_grids_binary = {
    'Logistic Regression': {
        'C': [0.1, 0.3, 0.5],
        'solver': ['lbfgs'],
        'max_iter': [500, 1000]
    },
    'Random Forest': {
        'n_estimators': [50, 100],
        'max_depth': [10, 30],
        'min_samples_leaf': [1],
        'min_samples_split': [10],
        'class_weight': [None],
    },
    'XGBoost': {
        'n_estimators': [100, 200],
        'learning_rate': [0.05, 0.1],
        'max_depth': [5, 7],
    }
}


# --- –ë–∞–∑–æ–≤—ã–µ –º–æ–¥–µ–ª–∏ ---
base_models = {
    'Logistic Regression': LogisticRegression(random_state=21, multi_class='multinomial'),
    'Random Forest': RandomForestClassifier(random_state=21, class_weight='balanced', n_jobs=2),
    'XGBoost': XGBClassifier(random_state=21, objective='multi:softprob', n_jobs=-1, eval_metric='mlogloss')
}

# --- 4. GridSearchCV ---
# –û–±–æ—Ä–∞—á–∏–≤–∞–µ–º —Ü–∏–∫–ª –≤ tqdm –¥–ª—è –æ—Ç–æ–±—Ä–∞–∂–µ–Ω–∏—è –ø—Ä–æ–≥—Ä–µ—Å—Å-–±–∞—Ä–∞
for name, model in tqdm(base_models.items(), desc="–û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–µ–π"):
    print(f"\n{'='*70}\n–ó–∞–ø—É—Å–∫ GridSearchCV –¥–ª—è {name}...")
    start_time = time.time()
    
    grid_search = GridSearchCV(
        estimator=model,
        param_grid=param_grids_5class[name],
        scoring='accuracy',
        cv=cv_strategy,
        n_jobs=-1,
        verbose=3, # –£–≤–µ–ª–∏—á–∏–≤–∞–µ–º —É—Ä–æ–≤–µ–Ω—å –¥–µ—Ç–∞–ª–∏–∑–∞—Ü–∏–∏ –¥–ª—è –ø—Ä–æ—Å–º–æ—Ç—Ä–∞ –ø—Ä–æ–≥—Ä–µ—Å—Å–∞
        return_train_score=True
    )
    grid_search.fit(X_train_5class, y_train_5class)

tuned_models = {}
grid_search_results = {}

for name, model in base_models.items():
    print(f"\n{'='*70}\n–ó–∞–ø—É—Å–∫ GridSearchCV –¥–ª—è {name}...")
    start_time = time.time()
    
    grid_search = GridSearchCV(
        estimator=model,
        param_grid=param_grids_5class[name],
        scoring='accuracy',
        cv=cv_strategy,
        n_jobs=-1,
        verbose=0,
        return_train_score=True
    )
    grid_search.fit(X_train_5class, y_train_5class)
    
    tuned_models[name] = grid_search.best_estimator_
    grid_search_results[name] = grid_search
    
    print(f"–õ—É—á—à–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –¥–ª—è {name}: {grid_search.best_params_}")
    print(f"–õ—É—á—à–∞—è CV accuracy: {grid_search.best_score_:.3f}")
    print(f"GridSearch –∑–∞–≤–µ—Ä—à–µ–Ω –∑–∞ {(time.time()-start_time)/60:.1f} –º–∏–Ω")


# --- 5. –û—Ü–µ–Ω–∫–∞ –Ω–∞ —Ç–µ—Å—Ç–µ ---
final_results = {}
target_names = [f'Rating {i}' for i in sorted(unique_classes)]

for name, model in tuned_models.items():
    y_pred = model.predict(X_test_5class)
    accuracy = accuracy_score(y_test_5class, y_pred)
    final_results[name] = {
        'model': model,
        'y_pred': y_pred,
        'accuracy': accuracy,
        'classification_report': classification_report(
            y_test_5class, y_pred, labels=sorted(unique_classes),
            target_names=target_names, zero_division=0
        ),
        'confusion_matrix': confusion_matrix(y_test_5class, y_pred)
    }
    print(f"\n{name} Test Accuracy: {accuracy:.3f}")
    print(final_results[name]['classification_report'])

# --- 6. –°—Ä–∞–≤–Ω–µ–Ω–∏–µ –º–æ–¥–µ–ª–µ–π —Å baseline ---
most_common_class = pd.Series(y_train_5class).value_counts().index[0]
naive_accuracy = (y_test_5class == most_common_class).mean()
print(f"Naive baseline (–∫–ª–∞—Å—Å {most_common_class}): {naive_accuracy:.3f}")

comparison_data = []
for name in tuned_models.keys():
    cv_score = grid_search_results[name].best_score_
    test_score = final_results[name]['accuracy']
    comparison_data.append({
        'Model': name,
        'CV Accuracy': cv_score,
        'Test Accuracy': test_score,
        'Naive baseline': naive_accuracy,
        'Improvement vs Baseline': test_score - naive_accuracy
    })
comparison_df = pd.DataFrame(comparison_data)
print("\n–°—Ä–∞–≤–Ω–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤:")
print(comparison_df.to_string(index=False))

# --- 7. –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è ---
plt.figure(figsize=(12,6))
x = np.arange(len(comparison_df))
width = 0.25

plt.bar(x - width, comparison_df['CV Accuracy'], width, label='CV Accuracy', color='lightblue')
plt.bar(x, comparison_df['Test Accuracy'], width, label='Test Accuracy', color='lightcoral')
plt.bar(x + width, comparison_df['Naive baseline'], width, label='Baseline', color='lightgreen', alpha=0.6)

plt.xticks(x, comparison_df['Model'])
plt.ylabel('Accuracy')
plt.title('–°—Ä–∞–≤–Ω–µ–Ω–∏–µ –º–æ–¥–µ–ª–µ–π (CV, Test, Baseline)')
plt.legend()
plt.grid(axis='y', alpha=0.3)

for i, row in comparison_df.iterrows():
    plt.text(i - width, row['CV Accuracy'] + 0.01, f"{row['CV Accuracy']:.3f}", ha='center')
    plt.text(i, row['Test Accuracy'] + 0.01, f"{row['Test Accuracy']:.3f}", ha='center')
    plt.text(i + width, row['Naive baseline'] + 0.01, f"{row['Naive baseline']:.3f}", ha='center')

plt.tight_layout()
plt.show()

# --- 8. –ú–∞—Ç—Ä–∏—Ü—ã –æ—à–∏–±–æ–∫ ---
fig, axes = plt.subplots(1, 3, figsize=(20,6))
fig.suptitle('–ú–∞—Ç—Ä–∏—Ü—ã –æ—à–∏–±–æ–∫ –¥–ª—è –≤—Å–µ—Ö –º–æ–¥–µ–ª–µ–π', fontsize=16)
for i, (name, result) in enumerate(final_results.items()):
    cm = result['confusion_matrix']
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=target_names)
    disp.plot(ax=axes[i], cmap='Blues', values_format='d')
    axes[i].set_title(f"{name}\nTest Accuracy: {result['accuracy']:.3f}")
    axes[i].tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()

print("\nGridSearchCV –∑–∞–≤–µ—Ä—à–µ–Ω —É—Å–ø–µ—à–Ω–æ!")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import GridSearchCV, StratifiedKFold, train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay, roc_auc_score, roc_curve
import time

# 1. –°–û–ó–î–ê–ï–ú –î–í–ê –ö–õ–ê–°–°–ê (–±–∏–Ω–∞—Ä–Ω–∞—è –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏—è)
print("–°–æ–∑–¥–∞–Ω–∏–µ –¥–≤—É—Ö –∫–ª–∞—Å—Å–æ–≤...")
threshold = 4.0
y_binary = (y >= threshold).astype(int)

# –ü—Ä–æ–≤–µ—Ä—è–µ–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ
class_counts = np.bincount(y_binary)
class_names = ['bad (< 4.0)', 'good (>= 4.0)']
print("–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–æ–≤:")
for i, (count, name) in enumerate(zip(class_counts, class_names)):
    percentage = count / len(y_binary) * 100
    print(f"{name}: {count} —Ä–µ—Ü–µ–ø—Ç–æ–≤ ({percentage:.2f}%)")

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
colors = ['#ff6b6b', '#4ecdc4']
plt.bar(class_names, class_counts, color=colors)
plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–æ–≤')
plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ä–µ—Ü–µ–ø—Ç–æ–≤')
plt.xticks(rotation=45)

plt.subplot(1, 2, 2)
plt.pie(class_counts, labels=class_names, autopct='%1.1f%%', colors=colors)
plt.title('–ü—Ä–æ—Ü–µ–Ω—Ç–Ω–æ–µ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ')
plt.tight_layout()
plt.show()

# 2. –†–ê–ó–î–ï–õ–Ø–ï–ú –î–ê–ù–ù–´–ï
print("\n–†–∞–∑–¥–µ–ª–µ–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö –Ω–∞ train/test...")
X_train_binary, X_test_binary, y_train_binary, y_test_binary = train_test_split(
    X, y_binary, test_size=0.2, random_state=21, stratify=y_binary
)

print(f"–†–∞–∑–º–µ—Ä —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏: {X_train_binary.shape}")
print(f"–†–∞–∑–º–µ—Ä —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∏: {X_test_binary.shape}")

# 3. –ù–ê–°–¢–†–û–ô–ö–ê GRIDSEARCH
print("\n–ù–∞—Å—Ç—Ä–æ–π–∫–∞ GridSearchCV –¥–ª—è –±–∏–Ω–∞—Ä–Ω–æ–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏...")
cv_strategy = StratifiedKFold(n_splits=5, shuffle=True, random_state=21)

'''
# –°–µ—Ç–∫–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –¥–ª—è –±–∏–Ω–∞—Ä–Ω–æ–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏
param_grids_binary = {
    'Logistic Regression': {
        'C': [0.01, 0.1, 1, 10],
        'solver': ['lbfgs', 'liblinear'],
        'max_iter': [1000, 2000]
    },
    'Random Forest': {
        'n_estimators': [50, 100, 200],
        'max_depth': [None, 10, 20, 30],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4],
        'class_weight': ['balanced', None]
    },
    'XGBoost': {
        'n_estimators': [100, 200, 300],
        'learning_rate': [0.01, 0.05, 0.1, 0.2],
        'max_depth': [3, 5, 7, 9],
        'subsample': [0.6, 0.8, 1.0],
        'colsample_bytree': [0.6, 0.8, 1.0],
        'gamma': [0, 0.1, 0.2],
        'reg_alpha': [0, 0.1, 0.5],
        'reg_lambda': [1, 1.5, 2],
        'scale_pos_weight': [len(y_train_binary[y_train_binary == 0]) / len(y_train_binary[y_train_binary == 1])]  # –ë–∞–ª–∞–Ω—Å–∏—Ä–æ–≤–∫–∞ –∫–ª–∞—Å—Å–æ–≤
    }
}

'''

# –£–ø—Ä–æ—â–µ–Ω–Ω—ã–µ —Å–µ—Ç–∫–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –¥–ª—è –±—ã—Å—Ç—Ä–æ–≥–æ –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è
param_grids_binary = {
    'Logistic Regression': {
        'C': [0.1, 1],
        'solver': ['lbfgs'],
        'max_iter': [1000]
    },
    'Random Forest': {
        'n_estimators': [50, 100],
        'max_depth': [10, 20],
    },
    'XGBoost': {
        'n_estimators': [100, 200],
        'learning_rate': [0.05, 0.1],
        'max_depth': [5, 7],
    }
}


# –ë–∞–∑–æ–≤—ã–µ –º–æ–¥–µ–ª–∏ –¥–ª—è –±–∏–Ω–∞—Ä–Ω–æ–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏
base_models_binary = {
    'Logistic Regression': LogisticRegression(
        random_state=21,
        n_jobs=-1
    ),
    'Random Forest': RandomForestClassifier(
        random_state=21,
        n_jobs=-1
    ),
    'XGBoost': XGBClassifier(
        random_state=21,
        objective='binary:logistic',  # –ë–∏–Ω–∞—Ä–Ω–∞—è –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏—è
        n_jobs=-1,
        eval_metric='logloss'
    )
}

# 4. –ó–ê–ü–£–°–ö GRIDSEARCH
tuned_models_binary = {}
grid_search_results_binary = {}

for name, model in base_models_binary.items():
    print(f"\n{'='*70}")
    print(f"–ó–∞–ø—É—Å–∫ GridSearchCV –¥–ª—è {name} (–±–∏–Ω–∞—Ä–Ω–∞—è –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏—è)...")
    
    start_time = time.time()
    
    grid_search = GridSearchCV(
        estimator=model,
        param_grid=param_grids_binary[name],
        scoring='roc_auc',  # –ò—Å–ø–æ–ª—å–∑—É–µ–º ROC-AUC –¥–ª—è –±–∏–Ω–∞—Ä–Ω–æ–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏
        cv=cv_strategy,
        n_jobs=-1,
        verbose=1,
        return_train_score=True
    )
    
    # –ó–∞–ø—É—Å–∫ –ø–æ–∏—Å–∫–∞
    grid_search.fit(X_train_binary, y_train_binary)
    
    end_time = time.time()
    print(f"GridSearch –∑–∞–≤–µ—Ä—à–µ–Ω –∑–∞ {(end_time - start_time)/60:.1f} –º–∏–Ω—É—Ç")
    
    # –°–æ—Ö—Ä–∞–Ω—è–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã
    tuned_models_binary[name] = grid_search.best_estimator_
    grid_search_results_binary[name] = grid_search
    
    # –í—ã–≤–æ–¥ –ª—É—á—à–∏—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤
    print(f"–õ—É—á—à–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –¥–ª—è {name}:")
    for param, value in grid_search.best_params_.items():
        print(f"  {param}: {value}")
    print(f"–õ—É—á—à–∏–π ROC-AUC (CV): {grid_search.best_score_:.3f}")

# 5. –û–¶–ï–ù–ö–ê –ù–ê –¢–ï–°–¢–û–í–´–• –î–ê–ù–ù–´–•
print(f"\n{'='*70}")
print("–û–¶–ï–ù–ö–ê –ù–ê–°–¢–†–û–ï–ù–ù–´–• –ú–û–î–ï–õ–ï–ô –ù–ê –¢–ï–°–¢–û–í–´–• –î–ê–ù–ù–´–•")

final_results_binary = {}
target_names_binary = ['bad (< 4.0)', 'good (>= 4.0)']

for name, model in tuned_models_binary.items():
    print(f"\n{'='*60}")
    print(f"–û—Ü–µ–Ω–∫–∞ –Ω–∞—Å—Ç—Ä–æ–µ–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏ {name}...")
    
    # –ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è
    y_pred = model.predict(X_test_binary)
    y_pred_proba = model.predict_proba(X_test_binary)[:, 1] if hasattr(model, 'predict_proba') else None
    
    # –†–∞—Å—á–µ—Ç –º–µ—Ç—Ä–∏–∫
    accuracy = accuracy_score(y_test_binary, y_pred)
    roc_auc = roc_auc_score(y_test_binary, y_pred_proba) if y_pred_proba is not None else None
    
    final_results_binary[name] = {
        'model': model,
        'y_pred': y_pred,
        'y_pred_proba': y_pred_proba,
        'accuracy': accuracy,
        'roc_auc': roc_auc,
        'classification_report': classification_report(
            y_test_binary, 
            y_pred, 
            target_names=target_names_binary,
            zero_division=0
        ),
        'confusion_matrix': confusion_matrix(y_test_binary, y_pred)
    }
    
    print(f"Test Accuracy: {accuracy:.3f}")
    if roc_auc is not None:
        print(f"Test ROC-AUC: {roc_auc:.3f}")
    print("\nClassification Report:")
    print(final_results_binary[name]['classification_report'])

# 6. –°–†–ê–í–ù–ï–ù–ò–ï –†–ï–ó–£–õ–¨–¢–ê–¢–û–í
print(f"\n{'='*70}")
print("–§–ò–ù–ê–õ–¨–ù–û–ï –°–†–ê–í–ù–ï–ù–ò–ï –ú–û–î–ï–õ–ï–ô (–ë–∏–Ω–∞—Ä–Ω–∞—è –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏—è)")

# === –ù–∞–∏–≤–Ω–∞—è baseline-–º–æ–¥–µ–ª—å (–≤—Å–µ–≥–¥–∞ –ø—Ä–µ–¥—Å–∫–∞–∑—ã–≤–∞—Ç—å —Å–∞–º—ã–π —á–∞—Å—Ç—ã–π –∫–ª–∞—Å—Å)
most_common_class = pd.Series(y_train_binary).value_counts().index[0]
naive_accuracy_binary = (y_test_binary == most_common_class).mean()
print(f"Naive baseline (–≤—Å–µ–≥–¥–∞ –ø—Ä–µ–¥—Å–∫–∞–∑—ã–≤–∞—Ç—å –∫–ª–∞—Å—Å {most_common_class}): {naive_accuracy_binary:.3f}")

comparison_data_binary = []
for name in tuned_models_binary.keys():
    cv_score = grid_search_results_binary[name].best_score_  # ROC-AUC
    test_accuracy = final_results_binary[name]['accuracy']
    test_roc_auc = final_results_binary[name]['roc_auc']
    
    comparison_data_binary.append({
        'Model': name,
        'CV ROC-AUC': cv_score,
        'Test Accuracy': test_accuracy,
        'Naive baseline': naive_accuracy_binary,
        'Improvement vs Baseline': test_accuracy - naive_accuracy_binary,
        'Test ROC-AUC': test_roc_auc
    })

comparison_df_binary = pd.DataFrame(comparison_data_binary)
print("\n–°—Ä–∞–≤–Ω–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ (–≤–∫–ª—é—á–∞—è baseline):")
print(comparison_df_binary.to_string(index=False))


# 7. ROC-–ö–†–ò–í–´–ï
plt.figure(figsize=(10, 8))
for name, results in final_results_binary.items():
    if results['y_pred_proba'] is not None:
        fpr, tpr, _ = roc_curve(y_test_binary, results['y_pred_proba'])
        plt.plot(fpr, tpr, label=f'{name} (AUC = {results["roc_auc"]:.3f})', linewidth=2)

plt.plot([0, 1], [0, 1], 'k--', label='Random (AUC = 0.500)')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves - Binary Classification')
plt.legend(loc='lower right')
plt.grid(alpha=0.3)
plt.show()

# 8. –ú–ê–¢–†–ò–¶–´ –û–®–ò–ë–û–ö
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
fig.suptitle('–ú–∞—Ç—Ä–∏—Ü—ã –æ—à–∏–±–æ–∫ - –ë–∏–Ω–∞—Ä–Ω–∞—è –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏—è', fontsize=16)

for i, (name, result) in enumerate(final_results_binary.items()):
    cm = result['confusion_matrix']
    disp = ConfusionMatrixDisplay(
        confusion_matrix=cm, 
        display_labels=target_names_binary
    )
    disp.plot(ax=axes[i], cmap='Blues', values_format='d')
    axes[i].set_title(f'{name}\nAccuracy: {result["accuracy"]:.3f}\nAUC: {result["roc_auc"]:.3f}')
    axes[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

# 9. –ê–ù–ê–õ–ò–ó –í–ê–ñ–ù–û–°–¢–ò –ü–†–ò–ó–ù–ê–ö–û–í –î–õ–Ø –õ–£–ß–®–ï–ô –ú–û–î–ï–õ–ò
best_model_name_binary = max(final_results_binary.items(), key=lambda x: x[1]['roc_auc'])[0]
best_model_binary = final_results_binary[best_model_name_binary]['model']

print(f"\n–õ—É—á—à–∞—è –º–æ–¥–µ–ª—å: {best_model_name_binary}")

if hasattr(best_model_binary, 'feature_importances_'):
    feature_importances = pd.DataFrame({
        'feature': X.columns,
        'importance': best_model_binary.feature_importances_
    }).sort_values('importance', ascending=False)
    
    print("Top-15 —Å–∞–º—ã—Ö –≤–∞–∂–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤:")
    print(feature_importances.head(15))
    
    # –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è
    plt.figure(figsize=(12, 8))
    sns.barplot(x='importance', y='feature', data=feature_importances.head(15))
    plt.title(f'Top-15 Features Importance - {best_model_name_binary}\n(–ë–∏–Ω–∞—Ä–Ω–∞—è –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏—è)')
    plt.tight_layout()
    plt.show()

print("\nGridSearchCV –¥–ª—è –±–∏–Ω–∞—Ä–Ω–æ–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ —É—Å–ø–µ—à–Ω–æ –∑–∞–≤–µ—Ä—à–µ–Ω!")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import GridSearchCV, StratifiedKFold, train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.preprocessing import LabelEncoder
import time

# 1. –°–û–ó–î–ê–ï–ú –¢–†–ò –ö–õ–ê–°–°–ê –° –ß–ï–¢–ö–ò–ú–ò –ü–û–†–û–ì–ê–ú–ò
print("–°–æ–∑–¥–∞–Ω–∏–µ —Ç—Ä–µ—Ö –∫–ª–∞—Å—Å–æ–≤ —Å –ø–æ—Ä–æ–≥–∞–º–∏ 3.5 –∏ 4.2...")
threshold_low = 3.5
threshold_high = 4.2

# –°–æ–∑–¥–∞–µ–º –∫–∞—Ç–µ–≥–æ—Ä–∏–∏
bins = [0, threshold_low, threshold_high, 5.1]  # 5.1 —á—Ç–æ–±—ã –≤–∫–ª—é—á–∏—Ç—å 5.0
labels = ['bad', 'so-so', 'great']
y_3class = pd.cut(y, bins=bins, labels=labels, include_lowest=True)

# –ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –≤ —á–∏—Å–ª–æ–≤–æ–π —Ñ–æ—Ä–º–∞—Ç
label_encoder = LabelEncoder()
y_3class_encoded = label_encoder.fit_transform(y_3class)

# –°–æ—Ö—Ä–∞–Ω—è–µ–º mapping
class_mapping = dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
print("Mapping –∫–ª–∞—Å—Å–æ–≤:", class_mapping)

# –ü—Ä–æ–≤–µ—Ä—è–µ–º —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ
class_counts = np.bincount(y_3class_encoded)
class_names = label_encoder.classes_
print("\n–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–æ–≤:")
for i, (count, name) in enumerate(zip(class_counts, class_names)):
    percentage = count / len(y_3class_encoded) * 100
    print(f"{name}: {count} —Ä–µ—Ü–µ–ø—Ç–æ–≤ ({percentage:.2f}%)")

# –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è
plt.figure(figsize=(12, 5))
colors = ['#ff6b6b', '#ffe66d', '#4ecdc4']  # red, yellow, green

plt.subplot(1, 2, 1)
plt.bar(class_names, class_counts, color=colors)
plt.title('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –∫–ª–∞—Å—Å–æ–≤')
plt.ylabel('–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —Ä–µ—Ü–µ–ø—Ç–æ–≤')
plt.xticks(rotation=45)

plt.subplot(1, 2, 2)
plt.pie(class_counts, labels=class_names, autopct='%1.1f%%', colors=colors)
plt.title('–ü—Ä–æ—Ü–µ–Ω—Ç–Ω–æ–µ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏–µ')
plt.tight_layout()
plt.show()

# 2. –†–ê–ó–î–ï–õ–Ø–ï–ú –î–ê–ù–ù–´–ï
print("\n–†–∞–∑–¥–µ–ª–µ–Ω–∏–µ –¥–∞–Ω–Ω—ã—Ö –Ω–∞ train/test...")
X_train_3class, X_test_3class, y_train_3class, y_test_3class = train_test_split(
    X, y_3class_encoded, test_size=0.2, random_state=21, stratify=y_3class_encoded
)

print(f"–†–∞–∑–º–µ—Ä —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏: {X_train_3class.shape}")
print(f"–†–∞–∑–º–µ—Ä —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–∏: {X_test_3class.shape}")

# 3. –ù–ê–°–¢–†–û–ô–ö–ê GRIDSEARCH
print("\n–ù–∞—Å—Ç—Ä–æ–π–∫–∞ GridSearchCV –¥–ª—è —Ç—Ä–µ—Ö–∫–ª–∞—Å—Å–æ–≤–æ–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏...")
cv_strategy = StratifiedKFold(n_splits=3, shuffle=True, random_state=21)

'''
# –°–µ—Ç–∫–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –¥–ª—è —Ç—Ä–µ—Ö–∫–ª–∞—Å—Å–æ–≤–æ–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏
param_grids_3class = {
    'Logistic Regression': {
        'C': [0.01, 0.1, 1, 10],
        'solver': ['lbfgs', 'saga'],
        'max_iter': [1000, 2000],
        'class_weight': ['balanced', None]
    },
    'Random Forest': {
        'n_estimators': [50, 100, 200],
        'max_depth': [None, 10, 20, 30],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4],
        'class_weight': ['balanced', None]
    },
    'XGBoost': {
        'n_estimators': [50, 100, 150],
        'learning_rate': [0.01, 0.03, 0.05],
        'max_depth': [5, 7, 9],
        'subsample': [0.6, 0.8, 1.0],
        'colsample_bytree': [0.6, 0.8, 1.0],
        'gamma': [0, 0.1, 0.2],
        'reg_alpha': [0, 0.1, 0.5],
        'reg_lambda': [1, 1.5, 2]
    }
}

'''
# –£–ø—Ä–æ—â–µ–Ω–Ω—ã–µ —Å–µ—Ç–∫–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤
param_grids_3class = {
    'Logistic Regression': {
        'C': [0.1, 1, 10],
        'solver': ['lbfgs'],
        'max_iter': [1000]
    },
    'Random Forest': {
        'n_estimators': [100, 200],
        'max_depth': [10, 20],
    },
    'XGBoost': {
        'n_estimators': [100, 200],
        'learning_rate': [0.05, 0.1],
        'max_depth': [5, 7],
    }
}

# –ë–∞–∑–æ–≤—ã–µ –º–æ–¥–µ–ª–∏ –¥–ª—è —Ç—Ä–µ—Ö–∫–ª–∞—Å—Å–æ–≤–æ–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏
base_models_3class = {
    'Logistic Regression': LogisticRegression(
        random_state=21,
        
        n_jobs=-1
    ),
    'Random Forest': RandomForestClassifier(
        random_state=21,
        n_jobs=-1
    ),
    'XGBoost': XGBClassifier(
        random_state=21,
        objective='multi:softprob',
        num_class=3,
        n_jobs=-1,
        eval_metric='mlogloss'
    )
}

# 4. –ó–ê–ü–£–°–ö GRIDSEARCH
tuned_models_3class = {}
grid_search_results_3class = {}

for name, model in base_models_3class.items():
    print(f"\n{'='*70}")
    print(f"–ó–∞–ø—É—Å–∫ GridSearchCV –¥–ª—è {name} (3 –∫–ª–∞—Å—Å–∞)...")
    
    start_time = time.time()
    
    grid_search = GridSearchCV(
        estimator=model,
        param_grid=param_grids_3class[name],
        scoring='accuracy',
        cv=cv_strategy,
        n_jobs=-1,
        verbose=1,
        return_train_score=True
    )
    
    # –ó–∞–ø—É—Å–∫ –ø–æ–∏—Å–∫–∞
    grid_search.fit(X_train_3class, y_train_3class)
    
    end_time = time.time()
    print(f"GridSearch –∑–∞–≤–µ—Ä—à–µ–Ω –∑–∞ {(end_time - start_time)/60:.1f} –º–∏–Ω—É—Ç")
    
    # –°–æ—Ö—Ä–∞–Ω—è–µ–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç—ã
    tuned_models_3class[name] = grid_search.best_estimator_
    grid_search_results_3class[name] = grid_search
    
    # –í—ã–≤–æ–¥ –ª—É—á—à–∏—Ö –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤
    print(f"–õ—É—á—à–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä—ã –¥–ª—è {name}:")
    for param, value in grid_search.best_params_.items():
        print(f"  {param}: {value}")
    print(f"–õ—É—á—à–∞—è accuracy (CV): {grid_search.best_score_:.3f}")

# 5. –û–¶–ï–ù–ö–ê –ù–ê –¢–ï–°–¢–û–í–´–• –î–ê–ù–ù–´–•
print(f"\n{'='*70}")
print("–û–¶–ï–ù–ö–ê –ù–ê–°–¢–†–û–ï–ù–ù–´–• –ú–û–î–ï–õ–ï–ô –ù–ê –¢–ï–°–¢–û–í–´–• –î–ê–ù–ù–´–•")

final_results_3class = {}

for name, model in tuned_models_3class.items():
    print(f"\n{'='*60}")
    print(f"–û—Ü–µ–Ω–∫–∞ –Ω–∞—Å—Ç—Ä–æ–µ–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏ {name}...")
    
    # –ü—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è
    y_pred = model.predict(X_test_3class)
    y_pred_named = label_encoder.inverse_transform(y_pred)
    y_test_named = label_encoder.inverse_transform(y_test_3class)
    
    # –†–∞—Å—á–µ—Ç –º–µ—Ç—Ä–∏–∫
    accuracy = accuracy_score(y_test_3class, y_pred)
    
    final_results_3class[name] = {
        'model': model,
        'y_pred': y_pred,
        'y_pred_named': y_pred_named,
        'accuracy': accuracy,
        'classification_report': classification_report(
            y_test_named, 
            y_pred_named, 
            target_names=class_names,
            zero_division=0
        ),
        'confusion_matrix': confusion_matrix(y_test_named, y_pred_named)
    }
    
    print(f"Test Accuracy: {accuracy:.3f}")
    print("\nClassification Report:")
    print(final_results_3class[name]['classification_report'])

# 6. –°–†–ê–í–ù–ï–ù–ò–ï –†–ï–ó–£–õ–¨–¢–ê–¢–û–í
print(f"\n{'='*70}")
print("–§–ò–ù–ê–õ–¨–ù–û–ï –°–†–ê–í–ù–ï–ù–ò–ï –ú–û–î–ï–õ–ï–ô (3 –∫–ª–∞—Å—Å–∞)")

# Naive baseline (–ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–µ —Å–∞–º–æ–≥–æ —á–∞—Å—Ç–æ–≥–æ –∫–ª–∞—Å—Å–∞)
most_common_class = pd.Series(y_train_3class).value_counts().index[0]
naive_accuracy = sum(y_test_3class == most_common_class) / len(y_test_3class)
print(f"Naive baseline (–≤—Å–µ–≥–¥–∞ –ø—Ä–µ–¥—Å–∫–∞–∑—ã–≤–∞—Ç—å —Å–∞–º—ã–π —á–∞—Å—Ç—ã–π –∫–ª–∞—Å—Å): {naive_accuracy:.3f}")

comparison_data_3class = []
for name in tuned_models_3class.keys():
    cv_score = grid_search_results_3class[name].best_score_
    test_score = final_results_3class[name]['accuracy']
    
    comparison_data_3class.append({
        'Model': name,
        'CV Accuracy': cv_score,
        'Test Accuracy': test_score,
        'Naive baseline': naive_accuracy,
        'Improvement vs Baseline': test_score - naive_accuracy
    })

comparison_df_3class = pd.DataFrame(comparison_data_3class)

print("\n–°—Ä–∞–≤–Ω–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ (–≤–∫–ª—é—á–∞—è baseline):")
print(comparison_df_3class.to_string(index=False))


# 7. –í–ò–ó–£–ê–õ–ò–ó–ê–¶–ò–Ø –†–ï–ó–£–õ–¨–¢–ê–¢–û–í
plt.figure(figsize=(12, 6))
models = comparison_df_3class['Model']
x = np.arange(len(models))
width = 0.35

plt.bar(x - width/2, comparison_df_3class['CV Accuracy'], width, 
        label='CV Accuracy', alpha=0.8, color='lightblue')
plt.bar(x + width/2, comparison_df_3class['Test Accuracy'], width, 
        label='Test Accuracy', alpha=0.8, color='lightcoral')
plt.axhline(y=naive_accuracy, color='red', linestyle='--', 
           label=f'Naive baseline ({naive_accuracy:.3f})')

plt.xlabel('–ú–æ–¥–µ–ª–∏')
plt.ylabel('Accuracy')
plt.title('–°—Ä–∞–≤–Ω–µ–Ω–∏–µ —Ç–æ—á–Ω–æ—Å—Ç–∏ –Ω–∞ CV –∏ —Ç–µ—Å—Ç–æ–≤—ã—Ö –¥–∞–Ω–Ω—ã—Ö (3 –∫–ª–∞—Å—Å–∞)')
plt.xticks(x, models)
plt.legend()
plt.grid(axis='y', alpha=0.3)

# –î–æ–±–∞–≤–ª–µ–Ω–∏–µ –∑–Ω–∞—á–µ–Ω–∏–π
for i, row in comparison_df_3class.iterrows():
    plt.text(i - width/2, row['CV Accuracy'] + 0.01, f"{row['CV Accuracy']:.3f}", ha='center')
    plt.text(i + width/2, row['Test Accuracy'] + 0.01, f"{row['Test Accuracy']:.3f}", ha='center')

plt.tight_layout()
plt.show()

# 8. –ú–ê–¢–†–ò–¶–´ –û–®–ò–ë–û–ö
fig, axes = plt.subplots(1, 3, figsize=(20, 6))
fig.suptitle('–ú–∞—Ç—Ä–∏—Ü—ã –æ—à–∏–±–æ–∫ - –¢—Ä–µ—Ö–∫–ª–∞—Å—Å–æ–≤–∞—è –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏—è', fontsize=16)

for i, (name, result) in enumerate(final_results_3class.items()):
    cm = result['confusion_matrix']
    disp = ConfusionMatrixDisplay(
        confusion_matrix=cm, 
        display_labels=class_names
    )
    disp.plot(ax=axes[i], cmap='Blues', values_format='d')
    axes[i].set_title(f'{name}\nAccuracy: {result["accuracy"]:.3f}')
    axes[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

# 9. –ê–ù–ê–õ–ò–ó –í–ê–ñ–ù–û–°–¢–ò –ü–†–ò–ó–ù–ê–ö–û–í –î–õ–Ø –õ–£–ß–®–ï–ô –ú–û–î–ï–õ–ò
best_model_name_3class = max(final_results_3class.items(), key=lambda x: x[1]['accuracy'])[0]
best_model_3class = final_results_3class[best_model_name_3class]['model']

print(f"\n–õ—É—á—à–∞—è –º–æ–¥–µ–ª—å: {best_model_name_3class}")

if hasattr(best_model_3class, 'feature_importances_'):
    feature_importances = pd.DataFrame({
        'feature': X.columns,
        'importance': best_model_3class.feature_importances_
    }).sort_values('importance', ascending=False)
    
    print("Top-15 —Å–∞–º—ã—Ö –≤–∞–∂–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤:")
    print(feature_importances.head(15))
    
    # –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è
    plt.figure(figsize=(12, 8))
    sns.barplot(x='importance', y='feature', data=feature_importances.head(15))
    plt.title(f'Top-15 Features Importance - {best_model_name_3class}\n(–¢—Ä–µ—Ö–∫–ª–∞—Å—Å–æ–≤–∞—è –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏—è)')
    plt.tight_layout()
    plt.show()

print("\nGridSearchCV –¥–ª—è —Ç—Ä–µ—Ö–∫–ª–∞—Å—Å–æ–≤–æ–π –∫–ª–∞—Å—Å–∏—Ñ–∏–∫–∞—Ü–∏–∏ —É—Å–ø–µ—à–Ω–æ –∑–∞–≤–µ—Ä—à–µ–Ω!")
```
