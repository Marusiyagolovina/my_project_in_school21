### Делим данные на обучающую и тестовую выборку

```python
# Делим данные
X = df.drop(columns=["rating"])
y = df["rating"]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=21, stratify=y
)
```

## Naive regressor (предсказание среднего рейтинга)

```python
# Наивный алгоритм 1: Предсказываем среднее значение
mean_rating = y_train.mean()
y_pred_mean = np.full_like(y_test, mean_rating)

# Наивный алгоритм 2: Предсказываем медиану
median_rating = y_train.median()
y_pred_median = np.full_like(y_test, median_rating)

# Наивный алгоритм 3: Предсказываем самое частое значение (моду)
mode_rating = y_train.mode()[0] if not y_train.mode().empty else y_train.median()
y_pred_mode = np.full_like(y_test, mode_rating)

# 5. Оценка качества наивных алгоритмов
def evaluate_predictions(y_true, y_pred, model_name):
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_true, y_pred)
    
    
    print(f"\n📈 {model_name}:")
    print(f"MAE: {mae:.4f}")
    print(f"MSE: {mse:.4f}")
    print(f"RMSE: {rmse:.4f}")
    print(f"R²: {r2:.4f}")
    
    return {'MAE': mae, 'MSE': mse, 'RMSE': rmse, 'R²': r2}

# Оцениваем все наивные алгоритмы
results = {}
results['Mean'] = evaluate_predictions(y_test, y_pred_mean, "Наивный алгоритм (Среднее)")
results['Median'] = evaluate_predictions(y_test, y_pred_median, "Наивный алгоритм (Медиана)")
results['Mode'] = evaluate_predictions(y_test, y_pred_mode, "Наивный алгоритм (Мода)")

# 7. Анализ лучшего наивного алгоритма
best_naive = min(results, key=lambda x: results[x]['RMSE'])
print(f"\n🎯 Лучший наивный алгоритм: {best_naive}")
print(f"RMSE: {results[best_naive]['RMSE']:.4f}")
print(f"MAE: {results[best_naive]['MAE']:.4f}")

# 8. Бенчмарк для сравнения с будущими моделями
print(f"\n📊 Бенчмарк - наивный алгоритм ({best_naive}):")
print(f"Средняя ошибка: ±{results[best_naive]['MAE']:.3f} балла")
print(f"Типичная ошибка: ±{results[best_naive]['RMSE']:.3f} балла")

# 9. Анализ важности - какие ингредиенты чаще встречаются у высоких рейтингов
high_rated = df[df['rating'] >= df['rating'].quantile(0.75)]  # Топ 25% рецептов
low_rated = df[df['rating'] <= df['rating'].quantile(0.25)]   # Низ 25% рецептов

# Сравнение частоты ингредиентов
ingredient_impact = {}
for col in X.columns:
    if col != 'rating':
        high_freq = high_rated[col].mean()
        low_freq = low_rated[col].mean()
        impact = high_freq - low_freq
        ingredient_impact[col] = impact

# Топ 10 ингредиентов с наибольшим влиянием на рейтинг
top_positive = sorted(ingredient_impact.items(), key=lambda x: x[1], reverse=True)[:10]
top_negative = sorted(ingredient_impact.items(), key=lambda x: x[1])[:10]

print(f"\n Топ 10 ингредиентов для высокого рейтинга:")
for ing, impact in top_positive:
    print(f"  {ing}: +{impact:.3f}")

print(f"\n Топ 10 ингредиентов для низкого рейтинга:")
for ing, impact in top_negative:
    print(f"  {ing}: {impact:.3f}")
```

#### Разберём результаты:
- Mean (среднее значение рейтинга) → лучший baseline:

    - MAE ≈ 0.895

    - RMSE ≈ 1.341

    - R² ≈ 0.0 (как и положено наивной модели).

- Медиана и мода хуже — выше RMSE, ниже объяснённая дисперсия.

**Значит, любая модель должна бить этот порог (MAE < 0.895 и RMSE < 1.341)**, иначе она хуже "угадывания среднего".

🍽️ **Ингредиенты и рейтинг**

- Положительное влияние: мясо, молочные продукты (сыр, молоко, сливки), колбасы, ягоды (клюква).

- Отрицательное влияние: овощи (капуста, кабачок, редис, брокколи, перец), рис, банан, тофу, йогурт.

Это уже **даёт гипотезу:**

- "сытные" рецепты (мясо+молочка) чаще получают высокие оценки;

- "здоровая еда" и простые гарниры → наоборот.

Видим, при 5 разных разбиениях средний R² останлся отрицательным, это доказывает, что проблема не в случайности, а в самой модели или данных.

- Шумные данные: возможно, рейтинги содержат много случайности (субъективность пользователей).

- Мало информативных признаков: если все фичи — просто 0/1, модель видит очень ограниченную картину.

## Настроим гиперпараметры через GridSearchCV
Это следующий шаг, чтобы выжать максимальную производительность из XGBoost на наших данных

```python
from xgboost import XGBRegressor
from sklearn.model_selection import GridSearchCV

# Базовая модель
xgb_model = XGBRegressor(
    objective="reg:squarederror",  # регрессия
    random_state=21,
    n_jobs=-1
)

# Сетка параметров
param_grid = {
    "n_estimators": [200, 500, None],
    "max_depth": [2, 3, 5],
    "learning_rate": [0.01, 0.05, 0.1],
    "subsample": [0.7, 0.9],
    "colsample_bytree": [0.5, 0.7, 0.9]
}

# GridSearchCV
grid_search = GridSearchCV(
    estimator=xgb_model,
    param_grid=param_grid,
    scoring="r2",      # метрика для оценки
    cv=3,              # 3-fold кросс-валидация
    verbose=2,
    n_jobs=-1
)

# Запуск
grid_search.fit(X_train, y_train)

print("Лучшие параметры:", grid_search.best_params_)
print("Лучший R² (CV):", grid_search.best_score_)

# Оценим на тесте
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)

# Вычисляем метрики и сохраняем в переменные
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)  # или math.sqrt(mse)
r2 = r2_score(y_test, y_pred)

# Печатаем результаты
print(f"MSE (test): {mse:.3f}")
print(f"RMSE (test): {rmse:.3f}")
print(f"R² (test): {r2:.3f}")
```

#### Интерпретация результатов настройки  GridSearch

1.  **Лучшие параметры:** Алгоритм выбрал достаточно "осторожные" и сложные настройки:
    *   `'n_estimators': 500` (Много деревьев)
    *   `'learning_rate': 0.05` (Низкая скорость обучения)
    *   `'max_depth': 5` (Умеренная глубина деревьев)
    *   `'colsample_bytree': 0.5`, `'subsample': 0.7` (Довольно агрессивный отбор признаков и данных для борьбы с переобучением)

    **Что это значит:** Для того чтобы уловить очень слабые и сложные зависимости в наших данных, модели потребовалось много (`500`) мелких шажков (`learning_rate=0.05`) с ограничением сложности каждого шага (`max_depth=5`). Параметры `subsample` и `colsample_bytree` меньше 1.0 говорят о том, что модель склонна к переобучению на этих данных, и ей пришлось себя ограничивать.

2.  **Лучший R² (CV): 0.0416**
    *   Это средний результат на кросс-валидации. Модель на тренировочных данных (с перебором всех параметров) смогла объяснить **~4.16%** дисперсии. Уже небольшой прогресс.

3.  **R² (test): 0.037**
    *   **Это важный результат.** На абсолютно новых, незнакомых данных (тестовой выборке) модель показывает качество **0.037**.

4. **RMSE: 1.316**
    *   **Это самый важный и практичный показатель в нашем случае.** Модель предсказывает рейтинг рецепта с средней ошибкой в 1.316 балла.  Так как рейтинг находится в диапазоне от 1 до 5, ошибка в 1.3 балла — это колоссально много. Простое предсказание среднего справляется лучше (RMSE=1.341)


### Ключевые выводы

1.  **Минимальное улучшение:** Качество модели на тесте (`0.037`) практически не отличается от того, что было до подбора параметров (`0.036`). **GridSearch не дал существенного прироста.** Это красноречиво говорит о том, что мы уперлись в "потолок" предсказательной силы, который заложен в самих данных (в наших бинарных признаках-ингредиентах). Даже самый продвинутый алгоритм (XGBoost) с тщательно подобранными настройками не может найти сколько-нибудь значимой зависимости. **Потенциал признаков исчерпан.**

### Что это значит и что делать дальше?

**Все дальнейшие усилия должны быть направлены не на алгоритмы, а на данные:**

1.  **Feature Engineering (Создание новых признаков):**
    *   **Группируйте ингредиенты.** Создайте признаки не "лимон", "лайм", "апельсин", а `цитрусовые=1`.
    *   **Считайте синергию.** Создайте бинарные признаки для известных кулинарных пар: `томат_и_базилик=1`, `курица_и_чеснок=1`.
    *   **Добавьте мета-информацию.** Самый действенный способ. Если есть возможность, добавьте: `время_приготовления`, `сложность`, `категория_блюда` (десерт, суп, салат).

2.  **Сбор новых данных:** Это самый эффективный, но и самый трудозатратный путь.

3.  **Смена задачи:** Возможно, стоит упростить задачу. Вместо точного предсказания рейтинга (**регрессия**) попробуйте **бинарную классификацию**: например, предсказывать, будет ли рецепт "популярным" (рейтинг >= 4.5) или "непопулярным" (рейтинг < 4.0). Иногда так моделировать проще.

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import GridSearchCV, StratifiedKFold, train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay
import time
import warnings

# Подавляем FutureWarning sklearn
warnings.filterwarnings("ignore", category=FutureWarning, module="sklearn")

# --- 1. Создание классов ---
print("Создание классов...")
y_rounded = np.round(y).astype(int)
unique_classes = np.unique(y_rounded)
print(f"Уникальные классы: {unique_classes} ({len(unique_classes)})")
print("\nРаспределение классов:")
for r in sorted(unique_classes):
    count = sum(y_rounded == r)
    print(f"Рейтинг {r}: {count} ({count/len(y_rounded)*100:.1f}%)")

# --- 2. Разделение данных ---
X_train_5class, X_test_5class, y_train_5class, y_test_5class = train_test_split(
    X, y_rounded, test_size=0.2, random_state=21, stratify=y_rounded
)
print(f"Размер train: {X_train_5class.shape}, test: {X_test_5class.shape}")

# --- 3. Настройка GridSearch ---
cv_strategy = StratifiedKFold(n_splits=3, shuffle=True, random_state=21)

'''
# Расширенные сетки параметров для 5 классов
param_grids_5class = {
    'Logistic Regression': {
        'C': [0.01, 0.1, 1, 10],
        'solver': ['lbfgs'],
        'max_iter': [1000, 2000]
    },
    'Random Forest': {
        'n_estimators': [50, 100, 200],
        'max_depth': [None, 10, 20, 30],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4],
        'class_weight': ['balanced', None]
    },
    'XGBoost': {
        'n_estimators': [100, 200, 300],
        'learning_rate': [0.01, 0.05, 0.1, 0.2],
        'max_depth': [3, 5, 7, 9],
        'subsample': [0.6, 0.8, 1.0],
        'colsample_bytree': [0.6, 0.8, 1.0],
        'gamma': [0, 0.1, 0.2],
        'reg_alpha': [0, 0.1, 0.5],
        'reg_lambda': [1, 1.5, 2]
        # scale_pos_weight убран для многоклассовой задачи
    }
}
'''

# Упрощенные сетки параметров для быстрого выполнения
param_grids_binary = {
    'Logistic Regression': {
        'C': [0.1, 0.3, 0.5],
        'solver': ['lbfgs'],
        'max_iter': [500, 1000]
    },
    'Random Forest': {
        'n_estimators': [50, 100],
        'max_depth': [10, 30],
        'min_samples_leaf': [1],
        'min_samples_split': [10],
        'class_weight': [None],
    },
    'XGBoost': {
        'n_estimators': [100, 200],
        'learning_rate': [0.05, 0.1],
        'max_depth': [5, 7],
    }
}


# --- Базовые модели ---
base_models = {
    'Logistic Regression': LogisticRegression(random_state=21, multi_class='multinomial'),
    'Random Forest': RandomForestClassifier(random_state=21, class_weight='balanced', n_jobs=2),
    'XGBoost': XGBClassifier(random_state=21, objective='multi:softprob', n_jobs=-1, eval_metric='mlogloss')
}

# --- 4. GridSearchCV ---
# Оборачиваем цикл в tqdm для отображения прогресс-бара
for name, model in tqdm(base_models.items(), desc="Обучение моделей"):
    print(f"\n{'='*70}\nЗапуск GridSearchCV для {name}...")
    start_time = time.time()
    
    grid_search = GridSearchCV(
        estimator=model,
        param_grid=param_grids_5class[name],
        scoring='accuracy',
        cv=cv_strategy,
        n_jobs=-1,
        verbose=3, # Увеличиваем уровень детализации для просмотра прогресса
        return_train_score=True
    )
    grid_search.fit(X_train_5class, y_train_5class)

tuned_models = {}
grid_search_results = {}

for name, model in base_models.items():
    print(f"\n{'='*70}\nЗапуск GridSearchCV для {name}...")
    start_time = time.time()
    
    grid_search = GridSearchCV(
        estimator=model,
        param_grid=param_grids_5class[name],
        scoring='accuracy',
        cv=cv_strategy,
        n_jobs=-1,
        verbose=0,
        return_train_score=True
    )
    grid_search.fit(X_train_5class, y_train_5class)
    
    tuned_models[name] = grid_search.best_estimator_
    grid_search_results[name] = grid_search
    
    print(f"Лучшие параметры для {name}: {grid_search.best_params_}")
    print(f"Лучшая CV accuracy: {grid_search.best_score_:.3f}")
    print(f"GridSearch завершен за {(time.time()-start_time)/60:.1f} мин")


# --- 5. Оценка на тесте ---
final_results = {}
target_names = [f'Rating {i}' for i in sorted(unique_classes)]

for name, model in tuned_models.items():
    y_pred = model.predict(X_test_5class)
    accuracy = accuracy_score(y_test_5class, y_pred)
    final_results[name] = {
        'model': model,
        'y_pred': y_pred,
        'accuracy': accuracy,
        'classification_report': classification_report(
            y_test_5class, y_pred, labels=sorted(unique_classes),
            target_names=target_names, zero_division=0
        ),
        'confusion_matrix': confusion_matrix(y_test_5class, y_pred)
    }
    print(f"\n{name} Test Accuracy: {accuracy:.3f}")
    print(final_results[name]['classification_report'])

# --- 6. Сравнение моделей с baseline ---
most_common_class = pd.Series(y_train_5class).value_counts().index[0]
naive_accuracy = (y_test_5class == most_common_class).mean()
print(f"Naive baseline (класс {most_common_class}): {naive_accuracy:.3f}")

comparison_data = []
for name in tuned_models.keys():
    cv_score = grid_search_results[name].best_score_
    test_score = final_results[name]['accuracy']
    comparison_data.append({
        'Model': name,
        'CV Accuracy': cv_score,
        'Test Accuracy': test_score,
        'Naive baseline': naive_accuracy,
        'Improvement vs Baseline': test_score - naive_accuracy
    })
comparison_df = pd.DataFrame(comparison_data)
print("\nСравнение результатов:")
print(comparison_df.to_string(index=False))

# --- 7. Визуализация ---
plt.figure(figsize=(12,6))
x = np.arange(len(comparison_df))
width = 0.25

plt.bar(x - width, comparison_df['CV Accuracy'], width, label='CV Accuracy', color='lightblue')
plt.bar(x, comparison_df['Test Accuracy'], width, label='Test Accuracy', color='lightcoral')
plt.bar(x + width, comparison_df['Naive baseline'], width, label='Baseline', color='lightgreen', alpha=0.6)

plt.xticks(x, comparison_df['Model'])
plt.ylabel('Accuracy')
plt.title('Сравнение моделей (CV, Test, Baseline)')
plt.legend()
plt.grid(axis='y', alpha=0.3)

for i, row in comparison_df.iterrows():
    plt.text(i - width, row['CV Accuracy'] + 0.01, f"{row['CV Accuracy']:.3f}", ha='center')
    plt.text(i, row['Test Accuracy'] + 0.01, f"{row['Test Accuracy']:.3f}", ha='center')
    plt.text(i + width, row['Naive baseline'] + 0.01, f"{row['Naive baseline']:.3f}", ha='center')

plt.tight_layout()
plt.show()

# --- 8. Матрицы ошибок ---
fig, axes = plt.subplots(1, 3, figsize=(20,6))
fig.suptitle('Матрицы ошибок для всех моделей', fontsize=16)
for i, (name, result) in enumerate(final_results.items()):
    cm = result['confusion_matrix']
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=target_names)
    disp.plot(ax=axes[i], cmap='Blues', values_format='d')
    axes[i].set_title(f"{name}\nTest Accuracy: {result['accuracy']:.3f}")
    axes[i].tick_params(axis='x', rotation=45)
plt.tight_layout()
plt.show()

print("\nGridSearchCV завершен успешно!")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import GridSearchCV, StratifiedKFold, train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay, roc_auc_score, roc_curve
import time

# 1. СОЗДАЕМ ДВА КЛАССА (бинарная классификация)
print("Создание двух классов...")
threshold = 4.0
y_binary = (y >= threshold).astype(int)

# Проверяем распределение
class_counts = np.bincount(y_binary)
class_names = ['bad (< 4.0)', 'good (>= 4.0)']
print("Распределение классов:")
for i, (count, name) in enumerate(zip(class_counts, class_names)):
    percentage = count / len(y_binary) * 100
    print(f"{name}: {count} рецептов ({percentage:.2f}%)")

# Визуализация распределения
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
colors = ['#ff6b6b', '#4ecdc4']
plt.bar(class_names, class_counts, color=colors)
plt.title('Распределение классов')
plt.ylabel('Количество рецептов')
plt.xticks(rotation=45)

plt.subplot(1, 2, 2)
plt.pie(class_counts, labels=class_names, autopct='%1.1f%%', colors=colors)
plt.title('Процентное соотношение')
plt.tight_layout()
plt.show()

# 2. РАЗДЕЛЯЕМ ДАННЫЕ
print("\nРазделение данных на train/test...")
X_train_binary, X_test_binary, y_train_binary, y_test_binary = train_test_split(
    X, y_binary, test_size=0.2, random_state=21, stratify=y_binary
)

print(f"Размер тренировочной выборки: {X_train_binary.shape}")
print(f"Размер тестовой выборки: {X_test_binary.shape}")

# 3. НАСТРОЙКА GRIDSEARCH
print("\nНастройка GridSearchCV для бинарной классификации...")
cv_strategy = StratifiedKFold(n_splits=5, shuffle=True, random_state=21)

'''
# Сетки параметров для бинарной классификации
param_grids_binary = {
    'Logistic Regression': {
        'C': [0.01, 0.1, 1, 10],
        'solver': ['lbfgs', 'liblinear'],
        'max_iter': [1000, 2000]
    },
    'Random Forest': {
        'n_estimators': [50, 100, 200],
        'max_depth': [None, 10, 20, 30],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4],
        'class_weight': ['balanced', None]
    },
    'XGBoost': {
        'n_estimators': [100, 200, 300],
        'learning_rate': [0.01, 0.05, 0.1, 0.2],
        'max_depth': [3, 5, 7, 9],
        'subsample': [0.6, 0.8, 1.0],
        'colsample_bytree': [0.6, 0.8, 1.0],
        'gamma': [0, 0.1, 0.2],
        'reg_alpha': [0, 0.1, 0.5],
        'reg_lambda': [1, 1.5, 2],
        'scale_pos_weight': [len(y_train_binary[y_train_binary == 0]) / len(y_train_binary[y_train_binary == 1])]  # Балансировка классов
    }
}

'''

# Упрощенные сетки параметров для быстрого выполнения
param_grids_binary = {
    'Logistic Regression': {
        'C': [0.1, 1],
        'solver': ['lbfgs'],
        'max_iter': [1000]
    },
    'Random Forest': {
        'n_estimators': [50, 100],
        'max_depth': [10, 20],
    },
    'XGBoost': {
        'n_estimators': [100, 200],
        'learning_rate': [0.05, 0.1],
        'max_depth': [5, 7],
    }
}


# Базовые модели для бинарной классификации
base_models_binary = {
    'Logistic Regression': LogisticRegression(
        random_state=21,
        n_jobs=-1
    ),
    'Random Forest': RandomForestClassifier(
        random_state=21,
        n_jobs=-1
    ),
    'XGBoost': XGBClassifier(
        random_state=21,
        objective='binary:logistic',  # Бинарная классификация
        n_jobs=-1,
        eval_metric='logloss'
    )
}

# 4. ЗАПУСК GRIDSEARCH
tuned_models_binary = {}
grid_search_results_binary = {}

for name, model in base_models_binary.items():
    print(f"\n{'='*70}")
    print(f"Запуск GridSearchCV для {name} (бинарная классификация)...")
    
    start_time = time.time()
    
    grid_search = GridSearchCV(
        estimator=model,
        param_grid=param_grids_binary[name],
        scoring='roc_auc',  # Используем ROC-AUC для бинарной классификации
        cv=cv_strategy,
        n_jobs=-1,
        verbose=1,
        return_train_score=True
    )
    
    # Запуск поиска
    grid_search.fit(X_train_binary, y_train_binary)
    
    end_time = time.time()
    print(f"GridSearch завершен за {(end_time - start_time)/60:.1f} минут")
    
    # Сохраняем результаты
    tuned_models_binary[name] = grid_search.best_estimator_
    grid_search_results_binary[name] = grid_search
    
    # Вывод лучших параметров
    print(f"Лучшие параметры для {name}:")
    for param, value in grid_search.best_params_.items():
        print(f"  {param}: {value}")
    print(f"Лучший ROC-AUC (CV): {grid_search.best_score_:.3f}")

# 5. ОЦЕНКА НА ТЕСТОВЫХ ДАННЫХ
print(f"\n{'='*70}")
print("ОЦЕНКА НАСТРОЕННЫХ МОДЕЛЕЙ НА ТЕСТОВЫХ ДАННЫХ")

final_results_binary = {}
target_names_binary = ['bad (< 4.0)', 'good (>= 4.0)']

for name, model in tuned_models_binary.items():
    print(f"\n{'='*60}")
    print(f"Оценка настроенной модели {name}...")
    
    # Предсказания
    y_pred = model.predict(X_test_binary)
    y_pred_proba = model.predict_proba(X_test_binary)[:, 1] if hasattr(model, 'predict_proba') else None
    
    # Расчет метрик
    accuracy = accuracy_score(y_test_binary, y_pred)
    roc_auc = roc_auc_score(y_test_binary, y_pred_proba) if y_pred_proba is not None else None
    
    final_results_binary[name] = {
        'model': model,
        'y_pred': y_pred,
        'y_pred_proba': y_pred_proba,
        'accuracy': accuracy,
        'roc_auc': roc_auc,
        'classification_report': classification_report(
            y_test_binary, 
            y_pred, 
            target_names=target_names_binary,
            zero_division=0
        ),
        'confusion_matrix': confusion_matrix(y_test_binary, y_pred)
    }
    
    print(f"Test Accuracy: {accuracy:.3f}")
    if roc_auc is not None:
        print(f"Test ROC-AUC: {roc_auc:.3f}")
    print("\nClassification Report:")
    print(final_results_binary[name]['classification_report'])

# 6. СРАВНЕНИЕ РЕЗУЛЬТАТОВ
print(f"\n{'='*70}")
print("ФИНАЛЬНОЕ СРАВНЕНИЕ МОДЕЛЕЙ (Бинарная классификация)")

# === Наивная baseline-модель (всегда предсказывать самый частый класс)
most_common_class = pd.Series(y_train_binary).value_counts().index[0]
naive_accuracy_binary = (y_test_binary == most_common_class).mean()
print(f"Naive baseline (всегда предсказывать класс {most_common_class}): {naive_accuracy_binary:.3f}")

comparison_data_binary = []
for name in tuned_models_binary.keys():
    cv_score = grid_search_results_binary[name].best_score_  # ROC-AUC
    test_accuracy = final_results_binary[name]['accuracy']
    test_roc_auc = final_results_binary[name]['roc_auc']
    
    comparison_data_binary.append({
        'Model': name,
        'CV ROC-AUC': cv_score,
        'Test Accuracy': test_accuracy,
        'Naive baseline': naive_accuracy_binary,
        'Improvement vs Baseline': test_accuracy - naive_accuracy_binary,
        'Test ROC-AUC': test_roc_auc
    })

comparison_df_binary = pd.DataFrame(comparison_data_binary)
print("\nСравнение результатов (включая baseline):")
print(comparison_df_binary.to_string(index=False))


# 7. ROC-КРИВЫЕ
plt.figure(figsize=(10, 8))
for name, results in final_results_binary.items():
    if results['y_pred_proba'] is not None:
        fpr, tpr, _ = roc_curve(y_test_binary, results['y_pred_proba'])
        plt.plot(fpr, tpr, label=f'{name} (AUC = {results["roc_auc"]:.3f})', linewidth=2)

plt.plot([0, 1], [0, 1], 'k--', label='Random (AUC = 0.500)')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curves - Binary Classification')
plt.legend(loc='lower right')
plt.grid(alpha=0.3)
plt.show()

# 8. МАТРИЦЫ ОШИБОК
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
fig.suptitle('Матрицы ошибок - Бинарная классификация', fontsize=16)

for i, (name, result) in enumerate(final_results_binary.items()):
    cm = result['confusion_matrix']
    disp = ConfusionMatrixDisplay(
        confusion_matrix=cm, 
        display_labels=target_names_binary
    )
    disp.plot(ax=axes[i], cmap='Blues', values_format='d')
    axes[i].set_title(f'{name}\nAccuracy: {result["accuracy"]:.3f}\nAUC: {result["roc_auc"]:.3f}')
    axes[i].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

# 9. АНАЛИЗ ВАЖНОСТИ ПРИЗНАКОВ ДЛЯ ЛУЧШЕЙ МОДЕЛИ
best_model_name_binary = max(final_results_binary.items(), key=lambda x: x[1]['roc_auc'])[0]
best_model_binary = final_results_binary[best_model_name_binary]['model']

print(f"\nЛучшая модель: {best_model_name_binary}")

if hasattr(best_model_binary, 'feature_importances_'):
    feature_importances = pd.DataFrame({
        'feature': X.columns,
        'importance': best_model_binary.feature_importances_
    }).sort_values('importance', ascending=False)
    
    print("Top-15 самых важных признаков:")
    print(feature_importances.head(15))
    
    # Визуализация
    plt.figure(figsize=(12, 8))
    sns.barplot(x='importance', y='feature', data=feature_importances.head(15))
    plt.title(f'Top-15 Features Importance - {best_model_name_binary}\n(Бинарная классификация)')
    plt.tight_layout()
    plt.show()

print("\nGridSearchCV для бинарной классификации успешно завершен!")
```

```python
session = requests.Session()
session.proxies.update(proxies)
nutrition_url = "https://api.nal.usda.gov/fdc/v1/foods/search"
nutrition_data = []

daily = pd.read_csv(
    "data/daily.tsv", sep="\t",
    converters={"Daily Value": lambda value: float(value.replace(",", ""))}
)

for ingredients_list in tqdm(ingredients_list, desc="Fetching nutrition data"):
    try:
        response = session.get(
            nutrition_url,
            params={"query": ingredients_list, "api_key": api_key, "pageSize": 3},
            timeout=10
        )
        if response.status_code != 200:
            print(f"Сервер вернул {response.status_code} для {ingredients_list}")
            continue

        nutrition_info = response.json()
        foods = nutrition_info.get("foods", [])
        if not foods:
            print(f"Нет данных для {ingredients_list}")
            continue

        # ищем подходящий продукт
        for food in foods:
            desc = food.get("description", "").lower()
            if ingredients_list.lower() not in desc:
                continue

            nutrients = food.get("foodNutrients", [])
            nutrient_dict = {"ingredient": ingredients_list}

            for nutrient in nutrients:
                name = nutrient.get("nutrientName", "").lower()
                value = nutrient.get("value")
                if value is None:
                    continue
                for _, row in daily.iterrows():
                    if row["Nutrient"].lower() in name:
                        try:
                            nutrient_dict[row["Nutrient"]] = round(value / row["Daily Value"] * 100)
                        except ZeroDivisionError:
                            nutrient_dict[row["Nutrient"]] = None
                        break

            nutrition_data.append(nutrient_dict)
            break  # нашли подходящий продукт, идём к следующему ингредиенту

    except Exception as e:
        print(f"Ошибка для {ingredients_list}: {e}")
        continue
```

```python
nutrition_df = pd.DataFrame(nutrition_data)
nutrition_df.to_csv("data/nutrition_facts.csv", index=False)
```
